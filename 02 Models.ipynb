#**3. Data transformation**

## **3.1 Label-Enconding**

```python
from sklearn.preprocessing import LabelEncoder

# Columns to label encode
columns_to_encode = ['school', 'sex', 'address', 'famsize', 'Pstatus',
                     'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup',
                     'famsup', 'paid', 'activities', 'nursery', 'higher',
                     'internet', 'romantic']

# Apply label encoding using a loop
for col in columns_to_encode:
    x[col] = x[col].astype('category')
    x[col + '_new'] = LabelEncoder().fit_transform(x[col])

# Drop original categorical columns
x.drop(columns_to_encode, axis=1, inplace=True)
```

```python
x["guardian_new"].value_counts()
```

```python
df["guardian"].value_counts()
```

##**3.2 OneHot-Encoding**

```python
# Columns to one-hot encode
columns_to_encode_onehot = ['Mjob_new', 'Fjob_new', 'reason_new', 'guardian_new']

# Apply one-hot encoding using a loop
for col in columns_to_encode_onehot:
    # Convert to category type
    x[col] = x[col].astype('category')
    # Perform one-hot encoding
    enc_data = pd.get_dummies(x[col], prefix=col+'OHE', dtype='int')
    # Join the one-hot encoded columns to the dataframe
    x = pd.concat([x, enc_data], axis=1)

# Drop the original categorical columns
x.drop(columns_to_encode_onehot, axis=1, inplace=True)
```

#**4. Modeling**

##**4.1 Hands-on model training and evaluation**

### Data splitting and model definition

```python
# Data splitting
x1 = x.to_numpy()
y1 = y.to_numpy()
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.3, random_state=37)

# Reshape y_train and y_test
y_train = y_train.ravel()
y_test = y_test.ravel()

# Define models to use
models = {
    "KNN": KNeighborsRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Support Vector Machine": SVR(),
    "Decision Tree": DecisionTreeRegressor()
}

# Dictionary to store RMSE scores
rmse_scores = {}
```

### Model training and cross-validation

```python
# Train the models and obtain evaluation metric
for name, model in models.items():
    model.fit(X_train, y_train)
    # Validate the results using cross-validation
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores[name] = np.sqrt(-scores.mean())

# Compare the results
print("RMSE scores:")
for name, score in rmse_scores.items():
    print(f"{name}: {score:.4f}")
```

Random Forest is the model with the best evaluation score using RMSE, however, we will need to play around more with the different models parameters before reaching a final veredict on the best model

### Hyper-parameter search

```python
# Define hyperparameter grids for each model
param_grids = {
    "KNN": {'n_neighbors': [3, 5, 7, 9]},
    "Random Forest": {'n_estimators': [50, 100, 200, 300]},
    "Support Vector Machine": {'C': [0.1, 0.5, 1, 10]},
    "Decision Tree": {'max_depth': [None, 3, 5, 10]}
}

# Dictionary to store best RMSE scores for each model
best_rmse_scores = {}
# List to store all RMSE scores for each hyperparameter configuration
all_rmse_scores = []

# Train the models with different hyperparameters and obtain evaluation metric
for name, param_grid in param_grids.items():
    if name == "KNN":
        model = KNeighborsRegressor()
    elif name == "Random Forest":
        model = RandomForestRegressor()
    elif name == "Support Vector Machine":
        model = SVR()
    elif name == "Decision Tree":
        model = DecisionTreeRegressor()

    # Define the GridSearchCV with the appropriate model and hyperparameter grid
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1)

    # Train the GridSearchCV object
    grid_search.fit(X_train, y_train)

    # Store the RMSE scores for each hyperparameter configuration
    for i, score in enumerate(np.sqrt(-grid_search.cv_results_['mean_test_score'])):
        all_rmse_scores.append({'Model': name, 'Hyperparameter': i+1, 'RMSE': score})

    # Store the best RMSE score for the model
    best_rmse_scores[name] = np.sqrt(-grid_search.best_score_)

    # Print the best hyperparameters found for the model
    print(f"Best hyperparameters for {name}: {grid_search.best_params_}")

all_rmse_scores_df = pd.DataFrame(all_rmse_scores)

# Plot the RMSE scores for each hyperparameter configuration
plt.figure(figsize=(12, 8))

# Loop through each model and plot the RMSE scores
for model_name, model_df in all_rmse_scores_df.groupby('Model'):
    plt.plot(model_df['Hyperparameter'], model_df['RMSE'], marker='o', label=model_name)

plt.title('Change in RMSE with Different Hyperparameters')
plt.xlabel('Hyperparameter Configuration')
plt.ylabel('RMSE')
plt.xticks(range(1, 5))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Compare the results
print("\n RMSE scores:\n")
for name, score in best_rmse_scores.items():
    print(f"{name}: {score:.4f}")
```
